```python
import sacrebleu
import evaluate
from bert_score import score

reference = """
Pitch bearing vibration on NW-023 trending upward.
Follow-up in 30 days.
CTV mooring line replaced on Havørn.
"""

prediction = """
Routine inspections at Hvide Sande.
NW-023 pitch bearing elevated vibration.
CTV mooring line replaced.
Downtime reduced by 23%.
"""

references = [reference]
predictions = [prediction]

print("=== BLEU ===")

# sacrebleu expects list of predictions and list of reference lists
bleu = sacrebleu.corpus_bleu(predictions, [references])

print("BLEU score:", round(bleu.score, 2))
print("Precisions:", bleu.precisions)
print("BP (brevity penalty):", round(bleu.bp, 3))

print("\n=== ROUGE ===")

rouge = evaluate.load("rouge")
rouge_results = rouge.compute(predictions=predictions, references=references)

print("ROUGE-1:", round(rouge_results["rouge1"], 4))
print("ROUGE-2:", round(rouge_results["rouge2"], 4))
print("ROUGE-L:", round(rouge_results["rougeL"], 4))

print("\n=== BERTScore ===")

P, R, F1 = score(predictions, references, lang="en", verbose=False)

print("Precision:", round(P.mean().item(), 4))
print("Recall:", round(R.mean().item(), 4))
print("F1:", round(F1.mean().item(), 4))
```
